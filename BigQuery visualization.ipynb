```python
%load_ext google.cloud.bigquery
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

```python
%%bigquery queried_df
SELECT
    timestamp,
    volume,
    average
FROM `tradesponder.blotter.ESZ9_334144679`
```

```python
df = queried_df
```

```python
df = df.set_index('timestamp')
df
```

```python
df[['average']].plot()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S', tz=df.index.tz))
```

```python
df[['volume']].plot()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S', tz=df.index.tz))
```
